```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("breast-cancer.csv")
```

```python
print(df["diagnosis"].value_counts(ascending=True),'\n\n')
```

```python
ax = plt.axes(projection="3d")
x_data = np.arange(-5,5,0.1)
y_data = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x_data,y_data)
Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X,Y,Z, cmap="plasma")
plt.show()
```

```python
plt.figure(figsize = (5,5))

labels = ["Malignant","Benign"]
explode = [0, 0.1]

plt.pie(df.diagnosis.value_counts(), labels = labels, explode = explode, autopct = '%1.1f%%')

plt.show()
```

```python
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=df)
plt.title('Breast Cancer Dataset - Scatter Plot')
plt.show()
```

```python

mean_features = ["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","symmetry_mean","fractal_dimension_mean"]

for i in mean_features:
    plt.figure(figsize = (15,7.5))
    sns.distplot(df[i], hist=True, kde=True,
                    bins=int(180/5), color = 'blue',
                     hist_kws={'edgecolor':'black'})
```

```python
plt.figure(figsize = (17, 15))

param_list = df.columns
sns.heatmap(df[param_list[1:]].corr(), annot = True, fmt = ".2f")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

algorithms = ['KNeighbors', 'Logistic Regression', 'SVR', 'Decision Tree', 'RandomForest', 'ANN', 'MLP']
accuracy_scores = [0.9766081871345029, 0.9707602339181286, 0.9824561403508771, 0.9064327485380117, 0.9473684210526315, 0.9707602339181286, 0.9736842105263158]

plt.pie(accuracy_scores, labels=algorithms, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Accuracy Scores of Different Algorithms')
plt.show()
```

```python
plt.plot(algorithms, accuracy_scores, marker='o', linestyle='-', color='green')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Algorithms')
plt.show()
```

```python
plt.violinplot(accuracy_scores, showmedians=True)
plt.xticks([1], ['Algorithms'])
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Algorithms')
plt.show()
```

```python
plt.fill_between(algorithms, accuracy_scores, color='orange', alpha=0.5)
plt.xlabel('Algorithms')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Algorithms')
plt.show()
```
